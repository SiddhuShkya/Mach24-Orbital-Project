### NDWI (Mean Over Time)

```python
img = imread(MEAN_OVER_TIME)
plt.figure(figsize=(14, 6))
plt.imshow(img)
plt.axis("off")
plt.title("NDWI Mean over Time (Existing Image)")
plt.show()
```

- **`Cyclical Moisture Pattern:`** NDWI shows strong annual seasonality, peaking in autumn (highest water content) and bottoming out in late winter/early spring (driest).

- **`Stable Extremes:`** Peak mean NDWI remains consistent at around 0.13. Trough values are consistently low at approximately 0.03.

- **`High Spatial Variability:`** The large shaded area indicates high uncertainty in water content across the area, particularly during the peak season.

- **`Dry Season Risk:`** The variability band dips into negative values during troughs, suggesting that some areas experience extremely low or no surface moisture at the driest time of year.

✅ NDWI peaks in pre-monsoon or wet months, lower in winter. Shows moisture availability trends.
