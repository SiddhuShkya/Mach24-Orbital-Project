### NDVI (Mean Over Time)

```python
img = imread(MEAN_OVER_TIME)
plt.figure(figsize=(14, 6))
plt.imshow(img)
plt.axis("off")  # Hide axes
plt.title("NDBI Mean over Time (Existing Image)")
plt.show()
```

- **`Inverse Seasonality:`** NDBI shows a clear annual cycle, peaking (highest built-up signal) in early spring and hitting a trough (lowest signal) in early autumn.

- **`Negative Values:`** All mean values are negative (0.03 to 0.13), suggesting the area is generally non-built-up or heavily vegetated.

- **`Extreme Variability:`** The wide range (from 0.04 to 0.20) indicates very high spatial heterogeneity (a mix of dense built-up areas and heavy vegetation) within the monitored region.

- **`Stable Pattern:`** The annual peak and trough magnitudes have remained consistent across all observed years.

 NDBI is negative overall (likely less urbanized areas). It slightly rises during dry months when vegetation declines.
