### NDVI (Mean Over Time)

```python
img = imread(MEAN_OVER_TIME)
plt.figure(figsize=(14, 6))
plt.imshow(img)
plt.axis("off")  # Hide axes
plt.title("NDVI Mean over Time (Existing Image)")
plt.show()
```

- **`Strong Seasonality:`** The NDVI shows a clear, repeating annual cycle, peaking in the late summer/autumn (vegetation peak) and hitting its lowest point in the winter/early spring (vegetation trough).

- **`Peak Stability:`** Maximum NDVI values have been stable around 0.30 across the cycles from 2022 to the projected peak in late 2024.

- **`High Variability:`** The wide shaded area indicates significant variability in vegetation health across the monitored area, with the highest uncertainty occurring during the peak growing season.

- **`Recent Status:`** The data ends with the NDVI undergoing its typical seasonal decline from the late 2024 peak toward the expected early 2025 trough.

✅ NDVI increases during spring and early summer (March–May) and decreases in winter (Dec–Feb), showing clear seasonality.
